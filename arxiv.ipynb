```python
import feedparser, random, time
import pandas as pd

def get_arxiv_keys():
    return ["title", "published", "id", "summary", "authors", "link"]


def search_arxiv(arxiv_id):
    url = f"http://export.arxiv.org/api/query?id_list={arxiv_id}"
    feed = feedparser.parse(url)
    entry = feed.entries[0]
    return (
        entry.title,
        entry.published,
        entry.id,
        entry.summary,
        entry.authors,
        entry.link,
    )
```

```python
df = pd.read_csv('scholar_data.csv')
```

```python
def get_arxiv_id(url:str|None)->str:
    if pd.isnull(url):
        return None
    if url.startswith('https://arxiv.org/abs/'):
        return url.split('/')[-1]
    elif url.startswith('https://arxiv.org/pdf/'):
        return url.split('/')[-1].replace('.pdf', '')
    else:
        return None
    
df['arxiv_id'] = df.Link.apply(get_arxiv_id)
```

```python
arxiv_data = {}
for arxiv_id in df.arxiv_id.unique():
    random_delay = random.uniform(0, 1) * 10
    print(f"Waiting for {random_delay} seconds before searching for arxiv id {arxiv_id}... ", end="")
    time.sleep(random_delay)
    arxiv_data[arxiv_id] = search_arxiv(arxiv_id)
```
