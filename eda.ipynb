```python
import pandas as pd
df = pd.read_csv('data/result.csv')
```

Categorize된 데이터는 300개

```python
df = df.loc[~df['Survey 대상인가'].isnull()]
df.index = range(0, 300)
df.shape[0]
```

```python
df['Survey 대상인가'].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
categories = df['Survey 대상인가'].unique()
cat2idx = {cat:idx for idx,cat in enumerate(categories)}
df['Survey 대상인가'] = df['Survey 대상인가'].map(cat2idx)
datapoints = df[['Survey 대상인가']].reset_index().values
```

```python
plt.figure(figsize=(10, 3))
plt.scatter(*(datapoints.T))
plt.show()
```
